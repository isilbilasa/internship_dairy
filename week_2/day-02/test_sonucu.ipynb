```python
from ultralytics import YOLO
#eğittiğim modeli drivendan çağırıyorum.
model = YOLO('market_dataset-2/YOLO_Egitim_Sonuclari/weights/best.pt')

#test edeceğim fotoğraf yolu.
test_file = 'market_dataset-2/images/test/549bec28-frame_0022.jpg'

results =model.predict(
    source=test_file,
    conf=0.5,
    show_labels = False,  #bu sınıf isimlerini kapatır.
    show_conf = False,  # confidence skorlarını kaldırır
    save=True
)

```
